### Extract the Datasets from the Resourse folder (consumer-and-government-spending-on-electric-cars-2015-2020)

```python
# Read the data into a panda DataFrame
spending_info_df=pd.read_csv('Resources/Hardik/consumer-and-government-spending-on-electric-cars-2015-2020.csv',skiprows=3)
spending_info_df.head()
```

```python
#renaming the dataframe
data_df=spending_info_df.rename(columns = {'Share of government spending as a percentage of total':'Government Share_%'})
```

```python
data_df.head()
```

```python
#renaming the dataframe
spending_info_renamed_df=data_df.rename(columns = {'Unnamed: 0':'Years'})
```

```python
spending_info_renamed_df.head()
```

```python
#data type info
spending_info_renamed_df.info()
```

```python
#changing Datatype of year column from int to object
spending_info_renamed_df['Years'] = pd.to_datetime(spending_info_renamed_df['Years'],format='%Y')
```

```python
spending_info_renamed_df.head()
```

```python
#data type info
spending_info_renamed_df.info()
```

```python
#Exporting cleaned CSV files to the resources folder
spending_info_renamed_df.to_csv('Resources/Hardik_Data/Consumer and Government spending cleaned.csv',index=False)
```

### Extract the Datasets from the Resourse folder (consumer-and-government-spending-on-electric-cars-2015-2020)

```python
# Read the data into a panda DataFrame
EV_stations_info_df=pd.read_csv('Resources/Hardik/Electric and Alternative Fuel Charging Stations.csv')
EV_stations_info_df.head()
```

```python
#Droping the unwanted columns
#byusing drop function of panda
EV_stations_info_cleaned_df_1=EV_stations_info_df.drop(['Plus4','Station Phone','Status Code','Expected Date','Groups With Access Code','Cards Accepted','BD Blends','NG Fill Type Code','NG PSI','EV Level1 EVSE Num','EV Level2 EVSE Num','EV DC Fast Count'],axis=1)
```

```python
#Droping the unwanted columns
EV_stations_info_cleaned_df_2=EV_stations_info_cleaned_df_1.drop(['EV Other Info','EV Network','EV Network Web','Owner Type Code','Federal Agency Name','Hydrogen Status Link','LPG Primary','E85 Blender Pump','EV Connector Types','Intersection Directions (French)','Access Days Time (French)','BD Blends (French)'],axis=1)
```

```python
EV_stations_info_cleaned_df=EV_stations_info_cleaned_df_2.drop(['Groups With Access Code (French)','Hydrogen Is Retail','Access Detail Code','Federal Agency Code','CNG Dispenser Num','CNG On-Site Renewable Source','CNG Total Compression Capacity','CNG Storage Capacity','LNG On-Site Renewable Source','E85 Other Ethanol Blends','EV Pricing','EV Pricing (French)','LPG Nozzle Types','Hydrogen Pressures','Hydrogen Standards','CNG Fill Type Code','CNG PSI','CNG Vehicle Class','LNG Vehicle Class','EV On-Site Renewable Source','Restricted Access'],axis=1)
```

```python
EV_stations_info_cleaned_df_final=EV_stations_info_cleaned_df.drop(['Federal Agency ID'],axis=1)
```

```python
EV_stations_info_cleaned_df_final.head()
```

```python
#Droping Duplicate Data
EV_stations_info_cleaned_df.drop_duplicates(keep=False, inplace=True)
```

```python
#Reoder the columns 
EV_Stations_df=EV_stations_info_cleaned_df[['ID','Fuel Type Code','City','State','ZIP','Latitude','Longitude','Country','Station Name','Street Address','Intersection Directions','Access Days Time','Geocode Status','Date Last Confirmed','Updated At','Open Date','NG Vehicle Class','Access Code','Facility Type']]
```

```python
#Final cleaned-ordered Database
EV_Stations_df.head()
```

```python
EV_Stations_df.info()
```

```python
#Exporting cleaned CSV files to the resources folder
EV_Stations_df.to_csv('Resources/Hardik_Data/US_EV_Stations_data.csv',index=False)
```
