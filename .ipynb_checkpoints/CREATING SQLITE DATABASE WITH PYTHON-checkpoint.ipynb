```python
# importing libraries
import pandas as pd
import glob
import os
  
```

```python
# merging the files
joined_files = os.path.join("Resources/Hardik_Data", "*.csv")
```

```python
# A list of all joined files is returned
joined_list = glob.glob(joined_files)
```

```python
# Finally, the files are joined
df = pd.concat(map(pd.read_csv, joined_list), ignore_index=True)
print(df)
```

```python
merged_df = pd.DataFrame(df)
```

```python
merged_df.head()
```

```python
#Exporting cleaned CSV files to the resources folder
merged_df.to_csv('Resources/Hardik_Data/Merged_Data.csv',index=False)
```
